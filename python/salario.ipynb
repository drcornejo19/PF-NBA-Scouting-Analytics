```python
import pandas as pd

# Leer el CSV de salarios usando ruta completa y raw string
salario = pd.read_csv(r"C:\Users\lucas\OneDrive\Escritorio\NBA EXPANSION\salaries_1985to2018.csv")

# Verificar que se cargó correctamente
print(salario.head())
```

```python
len(salario)
```

```python
import pandas as pd

# 1️⃣ Calcular el salario promedio por jugador en salario2
salario_mean = salario.groupby('player_id')['salary'].mean().reset_index()

# Renombrar la columna para mayor claridad
salario_mean = salario_mean.rename(columns={'salary': 'salario_mean'})

salario_mean.head()
```

```python
len(salario_mean)
```

```python
# Merge usando player_id y _id
players_salary = players_salary.merge(
    salario_mean,
    left_on='_id',
    right_on='player_id',
    how='left'
)

# Opcional: eliminar la columna 'player_id' si ya no la necesitas
if 'player_id' in players_salary.columns:
    players_salary = players_salary.drop(columns=['player_id'])

# Verificar
players_salary.head()
```

```python
len(players_salary)
```

```python
# Eliminar filas con NaN en la columna 'salario_mean'
players_salary = players_salary.dropna(subset=['salario_mean'])

# Verificar
print(f"Número de registros después de eliminar NaN: {len(players_salary)}")
players_salary.head()
```

```python
import pandas as pd

# Normalizar nombres: quitar espacios, pasar a minúsculas y eliminar tabulaciones
player_original['full_name_clean'] = player_original['full_name'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
players_salary['name_clean'] = players_salary['name'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)

# Encontrar coincidencias únicas
coincidencias = set(player_original['full_name_clean']).intersection(set(players_salary['name_clean']))

# Mostrar resultados
print(f"Número de coincidencias únicas entre player_original y players_salary: {len(coincidencias)}")
```

```python
import pandas as pd

# 1️⃣ Normalizar nombres en ambos datasets
player_original['full_name_clean'] = player_original['full_name'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)
players_salary['name_clean'] = players_salary['name'].str.strip().str.lower().str.replace(r'\s+', ' ', regex=True)

# 2️⃣ Filtrar solo columnas necesarias de players_salary y eliminar duplicados
salarios = players_salary[['name_clean', 'salario_mean']].drop_duplicates(subset='name_clean')

# 3️⃣ Hacer merge para agregar salario_mean a player_original
player_original = player_original.merge(
    salarios,
    left_on='full_name_clean',
    right_on='name_clean',
    how='left'
)

# 4️⃣ Renombrar la columna agregada y limpiar auxiliares
player_original = player_original.rename(columns={'salario_mean': 'salario_promedio'})
player_original = player_original.drop(columns=['full_name_clean', 'name_clean'])

# 5️⃣ Contar valores no NaN en la nueva columna
num_no_na = player_original['salario_promedio'].notna().sum()
total = len(player_original)
porcentaje_no_na = (num_no_na / total) * 100

print(f"Número de valores no NaN en 'salario_promedio': {num_no_na}")
print(f"Porcentaje de valores no NaN del total: {porcentaje_no_na:.2f}%")

# 6️⃣ Verificar las primeras filas
player_original.head()
```
