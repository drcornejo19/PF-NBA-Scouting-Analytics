```python
import pandas as pd
import os
import matplotlib.pyplot as plt

na_vals = ["NA", "N/A", "null", "NULL", ""]

# Calcular porcentaje de NA por CSV
porcentaje_na_por_csv = {}

for ruta_csv in csv_files:
    df = pd.read_csv(ruta_csv, na_values=na_vals)
    nombre_archivo = os.path.basename(ruta_csv)
    
    total_na = df.isna().sum().sum()
    total_celdas = df.size  # filas x columnas
    porcentaje_na = (total_na / total_celdas) * 100
    
    porcentaje_na_por_csv[nombre_archivo] = porcentaje_na

# Preparar datos para graficar (incluir play_by_play)
archivos = list(porcentaje_na_por_csv.keys())
porcentajes = list(porcentaje_na_por_csv.values())

# Crear grÃ¡fico de columnas
plt.figure(figsize=(14, 6))
plt.bar(archivos, porcentajes, color="orange")
plt.title("Porcentaje de valores NA por CSV ")
plt.xlabel("Archivo CSV")
plt.ylabel("% de NA")
plt.xticks(rotation=45, ha="right")

# Mostrar valores encima de cada barra
for i, v in enumerate(porcentajes):
    plt.text(i, v + 0.5, f"{v:.1f}%", ha="center", va="bottom", fontsize=9)

plt.tight_layout()
plt.show()

```

```python
import pandas as pd
import os

carpeta = r"C:\Users\lucas\OneDrive\Escritorio\Csv de nba"
na_vals = ["NA", "N/A", "null", "NULL", ""]

# Listar todos los CSV
csv_files = [os.path.join(carpeta, f) for f in os.listdir(carpeta) if f.endswith(".csv")]

# Diccionario para guardar resultados
resumen_por_csv = {}

for ruta_csv in csv_files:
    df = pd.read_csv(ruta_csv, na_values=na_vals)
    nombre_archivo = os.path.basename(ruta_csv)
    
    # 1ï¸âƒ£ Seleccionar columnas numÃ©ricas
    num_cols = df.select_dtypes(include=["number"]).columns
    
    # 2ï¸âƒ£ Excluir dummies (solo 0 y 1)
    num_cols_no_dummies = [
        col for col in num_cols
        if not df[col].dropna().isin([0,1]).all()
    ]
    
    # 3ï¸âƒ£ Excluir columnas con "id" o "year" en el nombre
    num_cols_final = [
        col for col in num_cols_no_dummies
        if ("id" not in col.lower()) and ("year" not in col.lower())
    ]
    
    # 4ï¸âƒ£ Crear DataFrame resumen
    resumen = pd.DataFrame(index=num_cols_final)
    if num_cols_final:  # solo si hay columnas para procesar
        resumen["media"] = df[num_cols_final].mean()
        resumen["mediana"] = df[num_cols_final].median()
        resumen["moda"] = df[num_cols_final].mode().iloc[0]
        resumen["desviacion_estandar"] = df[num_cols_final].std()
    else:
        resumen["media"] = []
        resumen["mediana"] = []
        resumen["moda"] = []
        resumen["desviacion_estandar"] = []
    
    # Guardar en diccionario
    resumen_por_csv[nombre_archivo] = resumen
    
    # Mostrar resumen de este CSV
    print(f"\n===== Resumen de {nombre_archivo} =====")
    display(resumen)
```

```python
import os
import pandas as pd
from IPython.display import display

# Diccionario para guardar resultados
resultados_duplicados = {}

for ruta_csv in csv_files:
    df = pd.read_csv(ruta_csv)
    
    # Contar duplicados por columna
    duplicados_por_columna = {col: df[col].duplicated(keep=False).sum() for col in df.columns}
    
    # Convertir a DataFrame (igual que hiciste con los NA)
    nombre_archivo = os.path.basename(ruta_csv)
    resultados_duplicados[nombre_archivo] = pd.DataFrame.from_dict(
        duplicados_por_columna, orient="index", columns=["cantidad_duplicados"]
    )

# Mostrar resultados
for archivo, df_dup in resultados_duplicados.items():
    print(f"\nðŸ“„ {archivo}")
    display(df_dup)
```
